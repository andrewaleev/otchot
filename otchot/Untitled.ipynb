```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('shipments_data.csv')
df
```

```python
# Группировка и агрегация
result = df.groupby('маршрут').agg({
    'вес_груза': ['mean', 'median', lambda x: pd.Series.mode(x).iloc[0] if not pd.Series.mode(x).empty else None],
    'объем_груза': ['mean', 'median', lambda x: pd.Series.mode(x).iloc[0] if not pd.Series.mode(x).empty else None],
    'доход': ['mean', 'median', lambda x: pd.Series.mode(x).iloc[0] if not pd.Series.mode(x).empty else None]
})

result.columns = ['_'.join(col).strip() for col in result.columns.values]
result.reset_index(inplace=True)

result
```

```python
df['дата_и_время_прибытия'] = pd.to_datetime(df['дата_и_время_прибытия'])

df['дата_и_время_прибытия'] = df['дата_и_время_прибытия'].dt.date
df
```

```python
df = df.drop_duplicates()
```

```python
columns_with_missing_values = ['доход', 'стоимость', 'тип_груза', 'маршрут']
mask = df[columns_with_missing_values].isna().any(axis=1)
rows_with_missing_values = df[mask]
df = df.dropna(subset=columns_with_missing_values)
```

```python
# Метод на основе стандартного отклонения
def find_outliers_sd(df, columns, threshold=3):
    outliers = pd.DataFrame()
    for column in columns:
        mean = df[column].mean()
        std_dev = df[column].std()
        upper_bound = mean + threshold * std_dev
        lower_bound = mean - threshold * std_dev
        outliers = pd.concat([outliers, df[(df[column] > upper_bound) | (df[column] < lower_bound)]])
    return outliers.drop_duplicates()

# Метод на основе межквартильного размаха (IQR)
def find_outliers_iqr(df, columns, factor=1.5):
    outliers = pd.DataFrame()
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        upper_bound = Q3 + factor * IQR
        lower_bound = Q1 - factor * IQR
        outliers = pd.concat([outliers, df[(df[column] > upper_bound) | (df[column] < lower_bound)]])
    return outliers.drop_duplicates()

# Определяем столбцы для проверки
numeric_columns = ['вес_груза', 'объем_груза', 'стоимость', 'доход']

# Поиск аномалий
outliers_sd = find_outliers_sd(df, numeric_columns)
outliers_iqr = find_outliers_iqr(df, numeric_columns)

print("Аномалии на основе стандартного отклонения:")
print(outliers_sd)

print("\nАномалии на основе межквартильного размаха (IQR):")
print(outliers_iqr)
```

```python
# Вычисление метрик с использованием медианы
metrics = {
    'Доход на отгрузку': round(df['доход'].mean(),2),
    'Стоимость на отгрузку': round(df['стоимость'].mean(),2),
    'Общая валовая прибыль': round(df['доход'].sum() - df['стоимость'].sum(),2),
    'Общая сумма продаж': round(df['доход'].sum(),2),
    'Коэффициент заполнения 1 контейнера': int(df['коэффициент_заполнения'].mean()*100),
    'Процент окупаемости контейнера':round(((df['доход'].mean() - (df['стоимость'].mean())) / (20_000_000)) * 100,2)
}
```

```python
# Печать результатов
print("Бизнес-метрики:")
for key, value in metrics.items():
    print(f"{key}: {value}")
```

```python
# Инициализация Dash
app = dash.Dash(__name__)

# Сортировка данных для графика по клиентам
sorted_df = df.groupby('имя_клиента')['доход'].sum().reset_index()
sorted_df = sorted_df.sort_values(by='доход', ascending=True)

formatted_metrics = {key: "{:,.2f}".format(value).replace(',', ' ') for key, value in metrics.items()}

# Создание графиков
fig_доход_по_дням = px.bar(df.groupby('дата_и_время_прибытия')['доход'].sum().reset_index(), x='дата_и_время_прибытия', y='доход', title='Сумма продаж за день')
fig_доход_по_маршрутам = px.pie(df.groupby('маршрут')['доход'].sum().reset_index(), names='маршрут', values='доход', title='Сумма продаж по маршруту')
fig_доход_по_клиентам = px.bar(sorted_df, y='имя_клиента', x='доход', title='Сумма продаж по клиентам')

# Определение макета дашборда
app.layout = html.Div([
    html.H1('Интерактивный Дашборд'),

    # Карточки метрик
    html.Div([
        html.Div([
            html.H3('Доход на отгрузку'),
            html.P(f'{formatted_metrics["Доход на отгрузку"]} ₽'),
        ], className='card'),

        html.Div([
            html.H3('Стоимость на отгрузку'),
            html.P(f'{formatted_metrics["Стоимость на отгрузку"]} ₽'),
        ], className='card'),

        html.Div([
            html.H3('Общая валовая прибыль'),
            html.P(f'{formatted_metrics["Общая валовая прибыль"]} ₽'),
        ], className='card'),

        html.Div([
            html.H3('Общая сумма продаж'),
            html.P(f'{formatted_metrics["Общая сумма продаж"]} ₽'),
        ], className='card'),

        html.Div([
            html.H3('Коэффициент заполнения 1 контейнера'),
            html.P(f'{formatted_metrics["Коэффициент заполнения 1 контейнера"]}'),
        ], className='card'),

        html.Div([
            html.H3('Процент окупаемости контейнера'),
            html.P(f'{formatted_metrics["Процент окупаемости контейнера"]}%'),
        ], className='card'),
    ], style={'display': 'flex', 'flexWrap': 'wrap', 'gap': '20px'}),
    
    # Графики
    dcc.Graph(figure=fig_доход_по_дням),
    dcc.Graph(figure=fig_доход_по_маршрутам),
    dcc.Graph(figure=fig_доход_по_клиентам),
], style={'display': 'flex', 'flexWrap': 'wrap', 'gap': '20px'})

# Запуск приложения
if __name__ == '__main__':
    app.run_server(debug=True)
```
